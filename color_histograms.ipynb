```python
def kde_sklearn(x, x_grid, bandwidth, **kwargs):
    from sklearn.neighbors import KernelDensity
    """Kernel Density Estimation with Scikit-learn"""
    kde_skl = KernelDensity(bandwidth=bandwidth, **kwargs)
    kde_skl.fit(x[:, np.newaxis])
    # score_samples() returns the log-likelihood of the samples
    log_pdf = kde_skl.score_samples(x_grid[:, np.newaxis])
    # Scott
    print 1.06*np.std(x)*x.shape[0]**(-0.2)
    return np.exp(log_pdf)
```

```python
pdf_red = kde_sklearn(image_rgb[:,:,0].flatten(), x_grid, bandwidth=0.1)
plt.plot(x_grid, pdf_red, color='red', alpha=0.5, lw=3)
plt.hold
plt.plot(x_grid, histogram_red,'r')
```

```python
pdf_green = kde_sklearn(image_rgb[:,:,1].flatten(), x_grid, bandwidth=0.1)
plt.plot(x_grid, pdf_green, color='green', alpha=0.5, lw=3)
plt.hold
plt.plot(x_grid, histogram_green,'g')
```

```python
pdf_blue = kde_sklearn(image_rgb[:,:,2].flatten(), x_grid, bandwidth=0.1)
plt.plot(x_grid, pdf_blue, color='blue', alpha=0.5, lw=3)
plt.hold
plt.plot(x_grid, histogram_blue,'b')
pdf_blue.getfield
```
