```python
import numpy as np
%matplotlib inline
```

```python
import matplotlib.pyplot as plt

from skimage import data, filter, color, exposure

import skimage.io as io

coll = io.ImageCollection('data/*.jpg')

# Load picture, convert to grayscale and detect edges
# 727, 742
image_rgb = coll[727]

# See the original picture
fig1, (ax1) = plt.subplots(ncols=1, nrows=1, figsize=(8, 4))

ax1.set_title('Original picture')
ax1.imshow(image_rgb)
ax1.hold
plt.show()
```

```python
bins = 256
histogram_red, _ = np.histogram(image_rgb[:,:,0].flatten(),bins, range=(0,255), normed=True)
histogram_green, _ = np.histogram(image_rgb[:,:,1].flatten(),bins,range=(0,255), normed=True)
histogram_blue, _ = np.histogram(image_rgb[:,:,2].flatten(),bins,range=(0,255), normed=True)
```

```python
plt.plot(range(0,bins), histogram_red,'r')
```

```python
plt.plot(range(0,bins), histogram_green,'g')
```

```python
plt.plot(range(0,bins), histogram_green,'b')
```

```python
# max_histogram_count = np.max([histogram_red, histogram_green, histogram_blue])
# histogram_red_normed = 1.0*histogram_red/max_histogram_count - 127
# histogram_green_normed = 1.0*histogram_green/max_histogram_count -127
# histogram_blue_normed = 1.0*histogram_blue/max_histogram_count -127
x_grid = np.arange(0,bins)
# print type(x_grid)
```

```python
def kde_sklearn(x, x_grid, bandwidth, **kwargs):
    from sklearn.neighbors import KernelDensity
    """Kernel Density Estimation with Scikit-learn"""
    kde_skl = KernelDensity(bandwidth=bandwidth, **kwargs)
    kde_skl.fit(x[:, np.newaxis])
    # score_samples() returns the log-likelihood of the samples
    log_pdf = kde_skl.score_samples(x_grid[:, np.newaxis])
    return np.exp(log_pdf)
```

```python
pdf = kde_sklearn(image_rgb[:,:,0].flatten(), x_grid, bandwidth=1)
plt.plot(x_grid, pdf, color='red', alpha=0.5, lw=3)
plt.hold
plt.plot(x_grid, histogram_red,'r')
```

```python
pdf = kde_sklearn(image_rgb[:,:,1].flatten(), x_grid, bandwidth=1)
plt.plot(x_grid, pdf, color='green', alpha=0.5, lw=3)
plt.hold
plt.plot(x_grid, histogram_green,'g')
```

```python
from sklearn.grid_search import GridSearchCV
grid = GridSearchCV(KernelDensity(),
                    {'bandwidth': np.linspace(0.1, 1.0, 30)},
                    cv=20) # 20-fold cross-validation
grid.fit(image_rgb[:,:,0].flatten()[:, None])
print grid.best_params_
```
