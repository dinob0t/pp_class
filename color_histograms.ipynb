```python
import numpy as np
%matplotlib inline
```

```python
import matplotlib.pyplot as plt

from skimage import data, filter, color, exposure

import skimage.io as io

coll = io.ImageCollection('data/*.jpg')

# Load picture, convert to grayscale and detect edges
# 727, 742
image_rgb = coll[727]

# See the original picture
fig1, (ax1) = plt.subplots(ncols=1, nrows=1, figsize=(8, 4))

ax1.set_title('Original picture')
ax1.imshow(image_rgb)
ax1.hold
plt.show()
```

```python
bins = 16
histogram_red, _ = np.histogram(image_rgb[:,:,0].flatten(),bins, range=(0,255))
histogram_green, _ = np.histogram(image_rgb[:,:,1].flatten(),bins,range=(0,255))
histogram_blue, _ = np.histogram(image_rgb[:,:,2].flatten(),bins,range=(0,255))
```

```python
plt.plot(range(0,16), histogram_red,'r')
```
